```python
from utils import CONFIG
from pathlib import Path
data_path = Path(CONFIG["tensorboard_data_path"] ) / "201_204"
import pandas as pd
import matplotlib.pyplot as plt
val_f1_dfs = []
metrics = ["val_acc_step", "val_accuracy", "val_f1_score"]
to_plot = {metric: [] for metric in metrics}
for metric in metrics:
        files = [a for a in (data_path / metric).iterdir() if a.is_file() and a.suffix == ".csv"]
        files.sort()
        for f in files:
            df = pd.read_csv(f)
            df.name = f.name
            to_plot[metric].append(df)

labels = [
    "tf_idf",
    "tf_idf + LaBSE",
    "count_vectorizer",
    "count_vectorizer + LaBSE",
    ]
figsize = (7, 5)
```

```python
for k,vs in to_plot.items():
    for df, label in zip(vs, labels):
        plt.plot(df["Step"], df["Value"], label=label)
    plt.legend()
    plt.ylabel(k.replace("_", " "))
    plt.xlabel("Step")
    plt.show()
```
