```python
from utils import get_artists, CONFIG
from tqdm import tqdm
from pathlib import Path
# get the word count for every artist and make a histogram
path_exper = CONFIG["experiments"]
d = {}
for art in tqdm(get_artists(path=path_exper)):
    d[art.name] = art.lyrics_len
d_sorted = {k:v for k,v in sorted(d.items(), key = lambda x: x[1], reverse=True)}
values = d.values()
d_sorted
```

```python
import matplotlib.pyplot as plt
no_zero = [v for v in values if v != 0]
print(no_zero)
print(len(no_zero))
print(len(values))
plt.hist(no_zero, bins=30, log=True)
plt.show()
more_2k = [v for v in values if v > 10000]
plt.hist(more_2k, bins=30, log=True)
plt.show()
plt.hist(more_2k[-50:], bins=30, log=True)
plt.show()
```
